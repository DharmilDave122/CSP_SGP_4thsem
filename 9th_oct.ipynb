```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://www.espncricinfo.com/series/icc-cricket-world-cup-2023-24-1367856/netherlands-vs-new-zealand-6th-match-1384397/full-scorecard"
```

```python
r = requests.get(url)
```

```python
#print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
table = soup.find("table" , class_ ="ds-w-full ds-table ds-table-md ds-table-auto  ci-scorecard-table")
```

```python
title = soup.find_all("th")
```

```python
print(title)
```

```python
header = []
```

```python
for i in title:
    name = i.text
    header.append(name)
```

```python
print(header)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=header)
```

```python
print(df)
```
